```python
import nsepy
from nsepy import get_history

from datetime import date
```

```python
import pandas as pd
```

```python
start_date = date(2017,3,1)
end_date = date.today()


tata_motors = nsepy.get_history(symbol="TataMotors", start= start_date, end= end_date)

tata_motors 
```

```python
tata_motors.to_csv(r'Tata_Motors_stock.csv')
```

```python
companies = {"M&M", "MARUTI", "ASHOKLEY", "EICHERMOT"}
```

```python
for company in companies:
    
    company_new = company.replace("&","%26")
    stock_data = nsepy.get_history(symbol=company_new, start=start_date, end=end_date)
    file_name = '{}.csv'.format(company)
    stock_data.to_csv(file_name, encoding= 'utf-8')
```
